```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
participant_data = pd.read_csv('data/cleaning data/participant_combined.csv')
covariate_data = ['eid','X34.0.0', 'Gender', 'Ethnicity', 'Migrant.Status', 'TDI.Tertiles', 'Highest.Qualification', 'House.Ownership', 'Income', 'Cohabiting', 'Living.Alone', 'AUDIT.Score', 'Smoker', 'Moderate.Physical.Activity', 'Longstanding.Illness', 'Diabetes', 'Cancer']
participant_data = participant_data[covariate_data]
participant_data.to_csv('data/cleaning data/participant_data.csv', index=False)
```

Gender

```python
gender_counts = participant_data['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(['Female', 'Male'], gender_counts, color=['lightpink', 'lightblue'])
plt.title('Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
plt.savefig('plots/participants/gender.png', dpi=300)
```

Ethnicity

```python
ethnicity_counts = participant_data['Ethnicity'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(ethnicity_counts.index, ethnicity_counts, color=['lightgray', 'lightblue', 'lightgreen', 'lightpink', 'lightyellow', 'lightcyan'])
plt.title('Distribution of Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.show()
plt.savefig('plots/participants/ethnicity.png', dpi=300)
```

Migrant Status

```python
migrant_counts = participant_data['Migrant.Status'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(['No', 'Yes'], gender_counts, color=['lightblue', 'lightgreen'])
plt.title('Migrant Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
plt.savefig('plots/participants/migrant status.png', dpi=300)
```

TDI Tertiles- Total Diet Index

```python
TDI_counts = participant_data['TDI.Tertiles'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(TDI_counts.index, TDI_counts, color=['lightgray', 'lightblue', 'lightgreen', 'lightpink', 'lightyellow', 'lightcyan'])
plt.title('TDI Tertiles')
plt.xlabel('Total Diet Index')
plt.ylabel('Count')
plt.show()
plt.savefig('plots/participants/TDI.png', dpi=300)
```

Income

```python
income_order = ['DontKnowRefuse', 'LessThan18K', '18Kto30K', '30Kto52K', '52Kto100K', 'MoreThan100K']
income_counts = participant_data['Income'].value_counts().loc[income_order]

plt.figure(figsize=(8, 6))
plt.bar(income_counts.index, income_counts, color=['lightgray', 'lightblue', 'lightgreen', 'lightpink', 'lightyellow', 'lightcyan'])
plt.title('Distribution of Income')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.show()
plt.savefig('plots/participants/income.png', dpi=300)
```

Cohabitating vs Living Alone

```python
alone_counts = participant_data['Living.Alone'].value_counts().loc[1]
cohab_counts = participant_data['Cohabiting'].value_counts().loc[1]

plt.figure(figsize=(8, 6))
plt.bar('Alone', alone_counts, color='lightblue')
plt.bar('Cohabitating', cohab_counts, color='lightgreen')
plt.title('Living Situation')
plt.xlabel('Living Situation')
plt.ylabel('Count')
plt.show()
plt.savefig('plots/participants/living situation.png', dpi=300)
```

```python
alcohol_counts = participant_data['AUDIT.Score'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(alcohol_counts.index, alcohol_counts, marker='o', linestyle='-', color='blue')
plt.title('Alcohol Use')
plt.xlabel('Alcohol Use')
plt.ylabel('Count')
plt.xlim(0, 20)
plt.grid(True)
plt.show()
plt.savefig('plots/participants/alcohol.png', dpi=300)
```
