```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression

# Load your dataframe
# df = pd.read_csv('your_data.csv')

# Sample dataframe for demonstration
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [1, 3, 2, 4, 5],
    'D': [5, 2, 3, 1, 4]
}
df = pd.DataFrame(data)

# Function to compute mutual information
def compute_mutual_information(df):
    mi = pd.DataFrame(index=df.columns, columns=df.columns)
    for col1 in df.columns:
        for col2 in df.columns:
            if df[col1].dtype == 'object' or df[col2].dtype == 'object':
                mi.loc[col1, col2] = mutual_info_classif(df[[col1]], df[col2])[0]
            else:
                mi.loc[col1, col2] = mutual_info_regression(df[[col1]], df[col2])[0]
    return mi.astype(float)

# Calculate mutual information
mi_matrix = compute_mutual_information(df)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(mi_matrix, annot=True, cmap='viridis')
plt.title('Mutual Information Heatmap')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

def anova_analysis(data, dependent_vars, significance_level=0.05):
    results = []
    significant_results = []

    # Fit ANOVA models for each dependent variable
    for var in dependent_vars:
        # Fit the ANOVA model
        model = ols(f'{var} ~ C(age_group)', data=data).fit()

        # Perform ANOVA
        anova_table = sm.stats.anova_lm(model, typ=2)
        
        # Extract p-value
        p_value = anova_table['PR(>F)'][0]

        # Append results to the list
        results.append({
            'Dependent Variable': var,
            'F-value': anova_table['F'][0],
            'p-value': p_value
        })
        
        # Check if the p-value is less than the significance level
        if p_value < significance_level:
            significant_results.append({
                'Dependent Variable': var,
                'p-value': p_value
            })

    # Create DataFrames from the results
    results_df = pd.DataFrame(results)
    significant_results_df = pd.DataFrame(significant_results)
    
    return results_df, significant_results_df


# Example usage
  # Replace with your actual DataFrame
dependent_variables = ['Sadness', 'Pessimism', 'Past_failure',
                       'Loss_of_pleasure', 'Guilty_feelings', 'Punishment_feelings',
                       'Self_dislike', 'Self_criticalness', 'Suicidal_thought_or_wishes',
                       'Crying', 'Agitation', 'Loss_of_interest', 'Indecisiveness',
                       'Worthlessness', 'Loss_of_energy', 'Changes_in_sleep_pattern',
                       'Irritability', 'Changes_in_appetite', 'Concentration_difficulty',
                       'Tiredness_or_fatigue', 'Loss_of_interest_in_sex', 'Systolic_blood_pressure',
                       'Diastolic_blood_pressure', 'Intima_media_thickness', 
                      'Alanine', 'Glutamine', 'Histidine', 'Isoleucine', 'Leucine',
       'Phenylalanine', 'Tyrosine', 'Valine', 'Apolipoprotein_A_I',
       'Apolipoprotein_B', 'Esterified_cholesterol', 'Free_cholesterol',
       'Total_cholesterol_HDL', 'Total_cholesterol_HDL2',
       'Total_cholesterol_HDL3', 'Total_cholesterol_LDL',
       'Remnant_cholesterol', 'Serum_total_cholesterol',
       'Total_cholesterol_VLDL', 'Linoleic_acid', 'Docosahexaenoic_acid',
       'Linoleic_acid_R18_2', 'Monounsaturated_fatty_acids',
       'Omega3_fatty_acids', 'Omega6_fatty_acids',
       'Polyunsaturated_fatty_acids', 'Saturated_fatty_acids',
       'Albumin_signal_area', 'Creatinine', 'Phosphatidylcholine',
       'Serum_total_triglycerides', 'Sphingomyelins', 'Total_cholines',
       'Total_phosphoglycerides', 'Triglycerides_HDL', 'Triglycerides_LDL',
       'Triglycerides_VLDL', 'Citrate', 'Glucose', 'Lactate',
       'Glycoprotein_acetyls', 'Hydroxybutyrate', 'Acetate', 'Acetoacetate',
       'HDL_particles_diameter', 'LDL_particles_diameter',
       'VLDL_particles_diameter', 'Degree_of_unsaturation',
       'Total_fatty_acids']

# Perform the ANOVA analysis
results_df, significant_results = anova_analysis(data_lim, dependent_variables)
significant_results
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set font size for better readability
plt.rcParams.update({'font.size': 12})

# Get the mean values before and after imputation
means_before_imputation = complete_stats['std']
means_after_imputation = imputed_stats['std']

# Create an array of labels for the statistics
statistics_labels = means_before_imputation.index

# Set the figure size and adjust spacing
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  # Width of the bars

# Plot bars for means before imputation
bars1 = ax.bar(np.arange(len(statistics_labels)), means_before_imputation, width, label='Raw data')

# Plot bars for means after imputation
bars2 = ax.bar(np.arange(len(statistics_labels)) + width, means_after_imputation, width, label='Imputed data')

# Add labels and title
ax.set_xlabel('Attributes with missing values')
ax.set_ylabel('Standard deviation')
ax.set_title('Comparison of Standard deviation Statistics Before and After Imputation')

ax.set_xticks(np.arange(len(statistics_labels)) + width / 2)
ax.set_xticklabels(statistics_labels, rotation=45, ha='right', fontsize=12)  



# Set logarithmic scale for y-axis
ax.set_yscale('log')

ax.legend()

plt.tight_layout()  

plt.show()
```
